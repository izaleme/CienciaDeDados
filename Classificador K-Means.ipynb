`# Encontrando o número ideal de clusters pelo método do cotovelo`

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#Iris Dataset
iris = datasets.load_iris()
X = iris.data
```

```python
# Coletando as distorções em uma lista
distortions = []
K = range(1,10)
for k in K:
  model = KMeans(n_clusters=k)
  model.fit(X)
  distortions.append(model.inertia_)
```

```python
# Plotando as distorções
#plt.figure(figsize=(12,6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção')
plt.title('O método do cotovelo (Elbow Method) mostrando o número ótimo de clusters')
plt.show()
```

`# Avaliando um modelo não supervisionado (utilizando o método da silhueta)`

```python
#importando as bibliotecas
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Gerando um dataset aleatório
A = np.random.rand(50,2)
B = 2 + np.random.rand(50,2)
X = np.concatenate((A,B))
```

```python
# Treinando o modelo e fazendo as predições
KMean= KMeans(n_clusters=2, random_state=123)
KMean.fit(X)
label=KMean.predict(X)
```

```python
# avaliando o modelo com o método da silhueta
print(f'Silhouette Score(n=2): {silhouette_score(X, label)}')
```

`# Importação das bibliotecas`

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets
from sklearn.metrics import silhouette_score
```

`# Carregando a base de dados Iris (https://archive.ics.uci.edu/ml/datasets/iris)`

```python
#Iris Dataset
iris = datasets.load_iris()
X = iris.data
```

`# Instanciando o classificador não supervisionado K-Means (https://scikit-learn.org/stable/modules/clustering.html#k-means)`

```python
km = KMeans(n_clusters=3)
```

`# Treinando o modelo`

```python
km.fit(X)
```

`# Fazendo a predição`

```python
km.predict(X)
labels = km.labels_

labels
```

`# Plotando um gráfico para mostrar o resultado`

```python
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
c=labels.astype(np.float), edgecolor="k", s=50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("K Means", fontsize=14)
```

```python
# avaliando o modelo com o método da silhueta
print(f'Silhouette Score: {silhouette_score(X, labels)}')
```

```python
km = KMeans(n_clusters=3)
km.fit(X)
```

`# Fazendo a predição`
